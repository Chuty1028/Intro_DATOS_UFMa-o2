```python
# Escribir archivo TXT de ejemplo
with open('datos.txt', 'w') as f:
    f.write('''Soy Alexander Lopez, producto musical
    Me gusto programar
    ''')

# Escribir archivo CSV de ejemplo
import csv
with open('datos.csv', 'w', newline='') as f:
    escritor = csv.writer(f)
    escritor.writerow(['Nombre', 'Edad', 'Ciudad'])
    escritor.writerow(['JUAN', 30, 'Guatemala'])
    escritor.writerow(['XIMENA', 29, 'Antigua'])
    escritor.writerow(['SERGIO', 19, 'Petén'])

print("Archivos de ejemplo creados")
```

```python
# Leer archivo TXT usando context manager
with open('datos.txt', 'r') as f:
    contenido = f.read()
    print("Contenido del archivo TXT:")
    print(contenido)

import csv

# Leer archivo CSV usando context manager
with open('datos.csv', 'r') as f:
    lector = csv.reader(f)
    print("Contenido del archivo CSV:")
    for fila in lector:
        print(fila)
```

**¿En qué parte o partes de un Pipeline de Datos se pueden usar las operaciones de Lectura (R) y Escritura (W)?**

En un pipeline de datos, las operaciones de Lectura (R) y Escritura (W) se usan en casi todas las etapas:

- Ingesta: se leen datos de fuentes y a veces se escriben en almacenamiento temporal.

- Transformación: se leen datos y se escriben ya procesados.

- Almacenamiento: principalmente se escriben los datos finales.

- Consumo: se leen los datos para análisis y a veces se escriben resultados.


La lectura y escritura ocurren continuamente a lo largo de todo el pipeline, ya que los datos se van moviendo y transformando entre etapas.



```python
import pandas as pd

# Crear un DataFrame
datos = {
    "Nombre": ["Juan", "Ana", "Luis"],
    "Edad": [25, 30, 28],
    "Ciudad": ["Guatemala", "Antigua", "Mixco"]
}

df = pd.DataFrame(datos)

# Guardar en CSV
df.to_csv("archivos_data_frame.csv", index=False, encoding="utf-8")
```

**¿Se podría usar un DataFrame para todos los tipos de datos (Estructurados, No Estructurados, Semi Estructurados)?**

Un DataFrame no se puede usar directamente para todos los tipos de datos.

Funciona muy bien con datos estructurados, también con datos semi estructurados después de procesarlos, pero no es adecuado para datos no estructurados sin antes transformarlos.


```python
# Escribir un archivo pickle

import pickle

datos = {
    "Nombre": ["Juan", "Ana", "Luis"],
    "Edad": [25, 30, 28]
}

# Guardar en archivo .pkl
with open("datos_pickle.pkl", "wb") as archivo:
    pickle.dump(datos, archivo)
```

```python
import pickle

# Cargar el archivo .pkl
with open("datos_pickle.pkl", "rb") as archivo:
    datos = pickle.load(archivo)

print(datos)
```

**¿Qué tipo de dato se podría almacenar con un Pickle?**


Pickle permite almacenar prácticamente cualquier objeto de Python en formato binario, incluyendo tipos básicos como enteros, strings y listas, estructuras como diccionarios y tuplas, y objetos más complejos como DataFrames de Pandas, modelos de machine learning o clases personalizadas; sin embargo, no es adecuado para objetos que dependen de recursos activos como archivos abiertos o conexiones.
