**Herramientas**

```python
# Herramientas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Extraer** Datos

```python
# Data set

df = pd.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")

print(df.shape)
df.head()
```

**Analisis Exploratorio de Datos (EDA)**

```python
# Tipo de datos
df.dtypes

# Información general
df.info()

df.isnull().sum()
```

**Analizar variables cuantitativas**

```python
# Estadísticas descriptivas
df.describe()
```

Grafico de Analisis Cuantitativo

```python
# Histograma
df['total_bill'].hist()
```

```python
# Boxplot
sns.boxplot(x=df['total_bill'])
```

```python
# Relación entre variables
sns.scatterplot(x='total_bill', y='tip', data=df)
```

# Analisis

**Analisis del histograma**

El histograma de la variable total_bill muestra una distribución sesgada hacia la derecha, donde la mayoría de las cuentas se concentran entre 10 y 25. Existen algunos valores altos que representan cuentas más elevadas, pero son menos frecuentes.

**Analisis de boxplot**

El diagrama de caja evidencia la presencia de valores atípicos (outliers) en la variable total_bill, correspondientes a cuentas considerablemente más altas que el promedio. La mediana se encuentra aproximadamente en valores intermedios, lo que indica una distribución relativamente equilibrada con algunos extremos.

**Analisis de Scatterplot**

El diagrama de dispersión entre total_bill y tip muestra una relación positiva, indicando que a medida que aumenta el valor de la cuenta, también aumenta la propina.
